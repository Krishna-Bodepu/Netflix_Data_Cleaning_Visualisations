### Data Visualisation on Netflix Dataset

```python
import pandas as pd
import numpy as np
```

#### Credits DataSet

```python
df = pd.read_csv('credits.csv')
df.head()
```

#### Titles Dataset

```python
df1 = pd.read_csv('titles.csv')
df1.head()
```

#### Merging the Two Datasets Credits and Titles baaed on id

```python
df = pd.merge(df,df1,on = 'id', how = 'inner')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

### Data Cleaning

#### Finding how many Null Values are there in dataset

```python
df.isna().sum()
```

#### Replacing NA values in Character with 'Not Specified'

```python
df["character"].fillna("Not Specified", inplace = True)
df.head()
```

#### Dropping the NA value in title

```python
df.dropna(subset=['title'], inplace = True)
df.head()
```

#### Replacing NA values in Description with 'No Description'

```python
df["description"].fillna("No Description", inplace = True)
df.head()
```

#### Filling NA values in age_certification with the method 'ffill'

```python
df['age_certification'].fillna(method='ffill', inplace=True)
df.head()
```

#### Replacing NA values in seasons with most frequent values

```python
df['seasons'].mode()
```

```python
df['seasons'].fillna(1,inplace = True)
df.head()
```

#### Replacing NA values in imdb_id with 'No IMDB_ID'

```python
df['imdb_id'].fillna('No IMDB_ID',inplace = True)
df.head()
```

#### Replacing NA values in imdb_score with the median value of imdb_score

```python
x = df['imdb_score'].median()
print(x)
```

```python
df['imdb_score'].fillna(x, inplace = True)
df.head()
```

#### Replacing NA values in imdb_votes with the summation of min and max value

```python
y = df['imdb_votes'].min() + df['imdb_votes'].max()
print(y)
```

```python
df['imdb_votes'].fillna(y, inplace = True)
df.head()
```

#### Dropping NA values rows in tmdb_popularity

```python
df.dropna(subset=['tmdb_popularity'], inplace = True)
df.head()
```

#### Filling NA values in tmdb_score with the method 'ffill'

```python
df['tmdb_score'].fillna(method='ffill', inplace=True)
df.head()
```

```python
df.isna().sum()
```

#### Getting unique values from the column 'genres'

```python
type(df['genres'][1])
```

```python
temp = list(set(df['genres'].values.tolist()))
temp
```

```python
l=[]
p=""
for i in range(len(temp)):
    x = temp[i]
    for j in x:
        if (j>='a' and j<='z') or (j>='A' and j<='Z'):
            p=p+j
        elif j==',' and p not in l:
            l.append(p)
            p=""
        elif p in l:
            p = ''
    if p not in l:
        l.append(p)
    p = ''
print(l)
print(len(l))
```

```python
l.remove('')
len(l)
```

#### Creating columns for the unique genres based on genres value in the dataset

```python
for i in range(len(l)):
    df[l[i]]=[1 if j.__contains__(l[i]) else 0 for j in df['genres']]
```
